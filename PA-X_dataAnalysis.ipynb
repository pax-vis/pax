# PA-X VISUALIZATION DATA ANALYSIS

```python
import pandas as pd
df = pd.read_csv("PAX_with_additional.csv")
df.head()
```

```python
print("Total Agreements", df.shape[0])
print("Codes per Agreement:",df.shape[1])
```

# Countries

```python
uniqueCon = df['Con'].unique()
print(uniqueCon)
print("Unique Countries & Country Combinations:",len(uniqueCon))
```

# Regions

```python
uniqueReg = df['Reg'].unique()
print(uniqueReg)
print("Unique Regions & Region Combinations:",len(uniqueReg))
allReg = df['Reg']
```

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# size the chart
fig = plt.figure(figsize=(20,20))
fig.add_subplot(1,1,1)
# plot the data
sns.countplot(x="Year", hue="Reg", data=df)
```

```python
# size the chart
#fig = plt.figure(figsize=(20,20))
#fig.add_subplot(1,1,1)
# style the chart - https://seaborn.pydata.org/tutorial/aesthetics.html
sns.set_style("ticks")  #darkgrid, whitegrid, dark, white, and ticks
# plot the data
sns.countplot(y="Reg", data=df)

allReg.value_counts()
```

# Years

```python
uniqueYears = df['Year'].unique()
print("Years:",uniqueYears)
print("Total Years:",len(uniqueYears))
print("Earliest Year:",uniqueYears.min())
print("Latest Year:",uniqueYears.max())
print(df['Year'].value_counts())
```

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(df['Year'], bins=26, kde=False, axlabel="Agreements per Year")
```

# Gender

```python
w_g_g = df['GeWom'] # women, girls, gender
m_b = df['GeMe'] # men, boys
g_s = df['GeSo'] # sexual orientation
g_f = df['GeFa'] # family
print("Women, girls and gender:")
print(w_g_g.value_counts())
print()
print("Men and boys:")
print(m_b.value_counts())
print()
print("Sexual orientation:")
print(g_s.value_counts())
print()
print("Family:")
print(g_f.value_counts())
print("Agreements that address gender:",315+85+9+170,"(",(100*((315+85+9+170)/1518.0)),"%)")
```

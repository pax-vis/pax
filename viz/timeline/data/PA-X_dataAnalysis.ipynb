```python
singleCons = []
for n in newCons:
    if ((n not in "RPMP/RPA/ABB") and ("RPMP/RPA/ABB" not in n)):
        splitCon = n.split("/")
        for s in splitCon:
            singleCons += [s]
    else:
        singleCons += [n]
singleCons.sort()
uniqueSingleCons = np.unique(singleCons)
print(uniqueSingleCons)
print(len(uniqueSingleCons))
```
